```python
StPer = pd.read_csv('https://stepik.org/media/attachments/course/4852/StudentsPerformance.csv')
```

```python
jj = [i for i in list(StPer) if 'score' in i]
StPer[jj]
```

```python
StPer.filter(like='score')
```

```python
mean_scores = StPer.groupby(['gender', 'race/ethnicity'], as_index = False) \
    .aggregate({'math_score': 'mean', 'writing_score': 'mean'}) \
    .rename(columns = {'math_score': 'mean_math_score', 'writing_score': 'mean_writing_score'})
```

```python
mean_scores
```

```python
StPer.groupby('gender', as_index = False) \
    .aggregate({'math_score': ['mean', 'count', 'std'], 'writing_score': ['mean', 'min', 'max']}) 
```

```python
StPer.sort_values(['gender', 'math_score'], ascending = False) \
    .groupby('gender').head()
```

```python
StPer['Total_score'] = StPer.filter(like='score').sum(axis=1)
```

```python
StPer = StPer.assign(tsl = np.log(StPer.Total_score))
StPer
```

```python
StPer.columns = StPer.columns.str.replace(' ','_')
StPer.reading_score.hist();
```

```python
StPer.plot.scatter(x='math_score', y='reading_score')
```

```python
ax = sns.lmplot(x='math_score', y='reading_score', hue='gender', data=StPer, fit_reg = False)
ax.set_xlabels('Math score')
ax.set_ylabels('Reading score')
```
